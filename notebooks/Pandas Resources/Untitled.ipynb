### Feature Engineering

Use domain knowledge to extract other important features -------
 This improves our ML algorithm

#### Fare per person

Fare per person would be fare paid by a person divided by family size

```python
data.columns
```

```python
data['Family_Size'] = data['SibSp'] + data['Parch'] + 1
data.head(1)
```

```python
data['Fare_Per_Person'] = data['Fare'] / data['Family_Size']
data.head(1)
```
